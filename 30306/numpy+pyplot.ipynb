```python
t=np.arange(0,12,1)
y1=np.sin(t)
y2=np.cos(t)
plt.plot(t,y1, 'r--', marker = '*')
plt.plot(t,y2, 'g--')

```

```python
plt.subplot(1, 2, 1)
plt.plot(t, y1, 'r--', marker='*')
plt.subplot(1, 2, 2)
plt.plot(t,y2, 'g--')
```

```python
from matplotlib import rc
rc('font', family='gulim')

x=np.arange(0, 5, 1)
y1 = x+1
y2 = x+2
y3 = x+3
y4 = x+4

plt.subplot(1, 2, 1)


plt.plot(x, y1, 'b--')
plt.plot(x, y3, 'r--')
plt.xlabel('mercy')
plt.ylabel('chara')
plt.legend(['undertale','amu'])


plt.subplot(1, 2, 2)

plt.plot(x, y2, 'y--', marker=',')
plt.plot(x, y4, 's--', marker='+')
plt.xlabel('mercy')
plt.ylabel('chara')
plt.legend(['undertale','amu'])

```
