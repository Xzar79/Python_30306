```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from PIL import Image
from wordcloud import WordCloud
img=r'C:\Users\user\Desktop\30306김환(워드클라우드)\태극기.png'
```

```python
Image.open(img)
```

```python
text = open(r'C:\Users\user\Desktop\30306김환(워드클라우드)\조광일한국.txt').read()
text
```

```python
lp_mask = np.array(Image.open(img))
lp_mask
```

```python
lp_wc = WordCloud(
    font_path = r'c:\Windows\Fonts\batang.ttc',
    background_color = 'white',
    mask=lp_mask,
    max_font_size=50,
    contour_width=5,
    contour_color='skyblue'
    ).generate(text)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```

```python
lp_wc.to_file(r'C:\Users\user\Desktop\30306김환(워드클라우드)\완성.png')
```
